### Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
data.columns
```

```python
def rating(rating):
  if rating>=7:
    return 'Excellent'
  elif rating>=6:
    return 'Good'
  else:
    return 'Average'
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```

### Find Unique Values From Genre

```python
data.columns
```

```python
data['Genre']
```

```python
list1=[]
for value in data ['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
  for value in item:
    one_d.append(value)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
  if item not in uni_list:
    uni_list.append(item)
```
