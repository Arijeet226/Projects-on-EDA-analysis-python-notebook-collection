chest pain type (4 values):

Value 0: typical angina

Value 1: atypical angina

Value 2: non-anginal pain

Value 3: asymptomatic

trestbps: resting blood pressure (in mm Hg on admission to the hospital)

chol: serum cholesterol in mg/dl

fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)

restecg: resting electrocardiographic results

Value 0: normal

Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression)

Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria

thalach: maximum heart rate achieved

exang: exercise induced angina (1 = yes; 0 = no)

oldpeak = ST depression induced by exercise relative to rest

slope: the slope of the peak exercise ST segment

Value 1: upsloping

Value 2: flat

Value 3: downsloping

ca: number of major vessels (0-3) colored by fluoroscopy

thal: 3 = normal; 6 = fixed defect; 7 = reversible defect

target: 0 = less chance of heart attack, 1 = more chance of heart attack

### Q-FIND COUNT OF MALE AND FEMALE

```python
data['sex'].value_counts()
```

```python
sns.countplot(x='sex',data=data,palette='Set2')
plt.xticks([0,1],['Female','Male'])
plt.show()
```

### Q- FIND GENDER DISTRIBUTION ACCORDING TO THE TARGET VARIABLE

```python
sns.countplot(x='sex',hue='target',data=data,palette='Set2')
plt.xticks([0,1],['Female','Male'])
plt.legend(['No Heart Disease','Heart Disease'])
plt.show()
```

### Q-CHECK AGE DISTRIBUTION

### Q-CHECK CHEST PAIN TYPE

### Q-SHOW CHEST PAIN DISTRIBUTION AS PER TARGET VARIABLE

```python
sns.countplot(x='cp',hue='target',data=data,palette='Set2')
plt.legend(labels=["No disease","Disease"])
plt.show()
```

### Q-SHOW FASTING BLOODSUGAR DISTRIBUTION ACCORDING TO TARGET VARIABLE

### Q-PLOT CONTINUOUS VARIABLES

```python
cate_val=[]
cont_val=[]

for column in data.columns:
  if data[column].nunique()<=10:
    cate_val.append(column)
  else:
    cont_val.append(column)
```
